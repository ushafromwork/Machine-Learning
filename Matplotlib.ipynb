Matplotlib:
1.Useful for making plots

```python
#import matplotlib library
import matplotlib.pyplot as plt
```

```python
import numpy as np
x= np.linspace(0,10,100)
y= np.sin(x)
z= np.cos(x)
```

```python
print(x)
```

```python
print(y)
print(z)
```

plotting the data

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
#adding title to x-axis and y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, 'r+')
plt.xlabel('angle')
plt.title('parabola')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x), 'r--')
plt.show()
```

```python
#Bar Plot
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','French','spanish','Latin','German']
people=[100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('No.of people')
plt.show()
```

```python
#Pie char
fig1= plt.figure()
ax1=fig1.add_axes([0,0,1,1])
languages=['English','French','spanish','Latin','German']
people=[100,50,150,40,70]
ax1.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

Scatter plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
axe2= fig2.add_axes([0,0,1,1])
axe2.scatter(x,y,color='g')
axe2.scatter(x,z,color='b')
plt.show()
```

3D scatter plot

```python
fig3= plt.figure()
ax=plt.axes(projection='3d')
z= 20 * np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```

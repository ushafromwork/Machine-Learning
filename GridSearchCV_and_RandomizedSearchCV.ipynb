**Hyper Parameter Tuning:**

1.GridSearchCV: We try all combination of parameters

2.RandomizedSearchCV: Randomly try few parameters with high accuracy

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
#We will be working on the breast cancer dataset
breast_cancer_dataset= sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#loading the data to a dataframe
data_frame=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
data_frame['label']=breast_cancer_dataset.target
```

```python
data_frame.head()
```

```python
data_frame.shape
```

```python
data_frame.isnull().sum()
```

```python
data_frame['label'].value_counts()
```

```python
x=data_frame.drop(columns='label',axis=1)
y=data_frame['label']
```

```python
print(y)
```

```python
x=np.asarray(x)
y=np.asarray(y)
```

#**GridSearchCV**

Grid SearchCV is used to determine the best parameters for our model

```python
#loading the SVC model
model=SVC()
```

```python
#HyperParameters
parameters={
             'kernel':['linear','poly','rbf','sigmoid'],
             'C':[1,5,10,20]
}
```

```python
#GridSearch CV
classifier=GridSearchCV(model,parameters,cv=5)#cv-cross validation,split data into different folds each time and test with different data
```

```python
#fitting data to our model
classifier.fit(x,y)
```

```python
classifier.cv_results_
```

```python
#Best Parameters
best_parameters=classifier.best_params_
print(best_parameters)
```

```python
#Determining highest accuracy
highest_accuracy=classifier.best_score_
print(highest_accuracy)
```

```python
#loading the results to panda dataframe
result=pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_serach_result=result[['param_C','param_kernel','mean_test_score']]
```

```python
print(grid_serach_result)
```

Highest Accuracy for SVM=95.2%
Best Parameters={'C':10,'kernel'='linear}

#**RandomizedSearchCV**

```python
model=SVC()
```

```python
#HyperParameters
parameters={
             'kernel':['linear','poly','rbf','sigmoid'],
             'C':[1,5,10,20]
}
```

```python
classifier=RandomizedSearchCV(model,parameters,cv=5)#cv-cross validation,split data into different folds each time and test with different data
```

```python
classifier.fit(x,y)
```

```python
classifier.cv_results_
```

```python
#Best Parameters
best_parameters=classifier.best_params_
print(best_parameters)
```

```python
#Determining highest accuracy
highest_accuracy=classifier.best_score_
print(highest_accuracy)
```

```python
#loading the results to panda dataframe
result=pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
randomized_serach_result=result[['param_C','param_kernel','mean_test_score']]
```

```python
randomized_serach_result
```

Highest Accuracy for SVM=95.2%

Best Parameters={'C':10,'kernel'='linear}

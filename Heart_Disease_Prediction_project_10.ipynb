```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

#Data Pre processing

```python
heart_data=pd.read_csv('/content/heart.csv')
heart_data.head()
```

```python
#last five rows
heart_data.tail()
```

```python
#NO.of rows and columns in data set
heart_data.shape
```

```python
#Getting some info about data
heart_data.info()
```

```python
#Checking for missing values
heart_data.isnull().sum()
```

```python
#Statistical measure about data
heart_data.describe()
```

```python
#Checking the distribution of target variable
# 1-- defective 0--not defective
heart_data['target'].value_counts()
```

Splitting features and target

```python
x=heart_data.drop(columns='target', axis=1)
y=heart_data['target']
```

```python
print(x)
```

```python
print(y)
```

#Splitting the data into training and test data

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

#Model Training

```python
#For Binary classification we use Logistic Regression model
model=LogisticRegression()
```

```python
#Training the model with training data
```

```python
model.fit(x_train,y_train)
```

#Model Evaluation
#Using Accuracy score

```python
#Accuracy on training data
x_tain_prediction= model.predict(x_train)
train_data_accuracy= accuracy_score(x_tain_prediction,y_train)
```

```python
print("Training data accuracy:", train_data_accuracy)
```

```python
#Accuracy on testdata
x_test_prediction= model.predict(x_test)
test_data_accuracy= accuracy_score(x_test_prediction,y_test)
```

```python
print("Test data accuracy:", test_data_accuracy)
```

```python
#Building predictive system
input_data=[46,1,0,120,249,0,0,144,0,0.8,2,0,3]

#change the input data to numpy array
input_data_as_array=np.asarray(input_data)

# Reshape the numpy array as we are predicting for data point
input_data_reshape=input_data_as_array.reshape(1,-1)

prediction= model.predict(input_data_reshape)
```

```python
print(prediction)
```

```python
if(prediction[0]==0):
  print(" The person doesn't have heart disease")
else:
  print("the person has heart disease")
```

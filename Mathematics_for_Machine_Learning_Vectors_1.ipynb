```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set() #it gives default theme
```

#Plotting vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1) #0,0 base pint of vector
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='Blue') #0,0 base pint of vector
plt.quiver(0,0,-3,-6,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

#Addition of vectors

```python
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])
sum=vector_1+vector_2
print(sum)
```

#Plotting the above vectors

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='Blue') #0,0 base pint of vector
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

#Subtraction of 2 vectors

```python
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])
sub=vector_1-vector_2
print(sub)
```

#Plotting subtraction of vectors

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='Blue') #0,0 base pint of vector
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,-1,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

#Multiplying a vector by a scalar

```python
vector_1=np.asarray([0,0,2,3])
mul=2*vector_1
print(mul)
```

#Plotting multiplication of vector by a scalar

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='Blue') #0,0 base pint of vector
plt.quiver(0,0,4,6,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1=np.asarray([0,0,2,3])
mul=-0.5*vector_1
print(mul)
```

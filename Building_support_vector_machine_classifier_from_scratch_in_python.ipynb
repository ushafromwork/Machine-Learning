Hyperplane:y=wx-b
Gradient Descent: w=w-learning_rate*dw
b=b-learning_rate*db

```python
import numpy as np
```

Support vector machine classifier

```python
class SVM_classifier():  #Class is a template

   # initiating the hyper parameters--parameters which we give manually
   def __init__(self,learning_rate,no_of_iterations,lambda_parameter):
    self.learning_rate=learning_rate
    self.no_of_iterations=no_of_iterations
    self.lambda_parameter=lambda_parameter

   # fitting the dataset to SVM Classifier
   def fit(self,x,y):
    #no.of datapoints in the data set(no.of rows)--m
    #no.of input features in the dataset(no.of columns)--n--no.of columns without target column
    self.m,self.n=x.shape

    #initiating weight and bias value ,no.of weights==no.of columns in dataset
    self.w =np.zeros(self.n)
    self.b=0
    self.x=x
    self.y=y

     #Implementing gradient descent
    for i in range(self.no_of_iterations):
      self.update_weights()


  #Function for updating weights
   def update_weights(self):
    y_label=np.where(self.y<= 0,-1,1)

    #gradients(dw,db)
    for index,x_i in enumerate(self.x):
      #gradients of gradient descent y_i(w.x_i -b)>1 and <1
      condition=y_label[index]*(np.dot(x_i,self.w)-self.b)>= 1
      if(condition==True):
        #dj/dw=2*lambda*w
          dw=2*self.lambda_parameter *self.w
          db=0

      else:
        dw= 2*self.lambda_parameter *self.w -np.dot(x_i,y_label[index])
        db=y_label[index]

      self.w=self.w-self.learning_rate * dw
      self.b=self.b-self.learning_rate * db


   def predict(self,x):

    output=np.dot(x,self.w)-self.b

    predicted_label=np.sign(output)

    y_hat=np.where(predicted_label <= -1,0,1)

    return y_hat
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
diabetes_data=pd.read_csv('/content/diabetes.csv')
```

```python
x=diabetes_data.drop(columns='Outcome',axis=1)
y=diabetes_data['Outcome']
```

```python
print(y)
```

```python
scalar=StandardScaler()
scalar.fit(x)
```

```python
standardized_data=scalar.transform(x)
print(standardized_data)
```

```python
x=standardized_data
y=diabetes_data['Outcome']
```

```python
#Train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Training the model

```python
classifier=SVM_classifier(learning_rate=0.001,no_of_iterations=1000,lambda_parameter=0.01)
```

```python
classifier.fit(x_train,y_train)
```

Model Evaluation

```python
#Accuracy on training data
x_train_prediction=classifier.predict(x_train)
accuracy_on_training_data=accuracy_score(y_train,x_train_prediction)
print(accuracy_on_training_data)
```

```python
#Accuracy on test data
x_test_prediction=classifier.predict(x_test)
accuracy_on_test_data=accuracy_score(y_test,x_test_prediction)
print(accuracy_on_test_data)
```

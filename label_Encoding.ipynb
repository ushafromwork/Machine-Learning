Label Encoding: predicting a datapoint belongs to one or the other

converting the labels into numeric form




```python
#importing the dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer dataset

```python
#loading the data from csv file to pandas dataframe
cancer_data=pd.read_csv('/content/diabetes.csv')
cancer_data.head()
```

```python
#finding the count of different labels
cancer_data['Outcome'].value_counts()
```

```python
#load the label encoder function
label_encoder=LabelEncoder()
```

```python
labels=label_encoder.fit_transform(cancer_data.Outcome)
```

```python
#Appending the labels to the dataframe
cancer_data['target']=labels

cancer_data.head()
```

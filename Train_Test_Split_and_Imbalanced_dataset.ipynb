```python
#Imbalanced dataset
credit_dataset=pd.read_csv('/content/credit_data.csv')
#first 5 rows of dataframe
credit_dataset.head()
```

```python
#last 5 rows of dataframe
credit_dataset.tail()
```

```python
#Distribution of 2 classes
credit_dataset['Class'].value_counts()
```

This is highly imbalanced datset
0---->legit(legal) transactions
1---->fraudulent

```python
#Separating the legit and fraudulent transactions
legit=credit_dataset[credit_dataset.Class==0]
fraud=credit_dataset[credit_dataset.Class==1]
print(legit)
print (fraud)
```

Under sampling--- taking fraudulent number of transactions from legit

Build a sample dataset containing similar distribution of legit and fraudulent,
fraudulent--492


```python
legit_sample=legit.sample(n=492)

print(legit_sample.shape)
```

concatenate the 2 dataframes, axis=0 means concatening row wise

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```
